# 🧠 Single Agent Pipeline Project

## Problem Statement
Build a **Single-Agent Smart Assistant** that:
- Understands user queries
- Routes tasks based on intent
- Uses tools when required
- Returns structured JSON output

### The agent should handle:
- Math queries → Calculator Tool
- Keyword extraction → Keyword Tool
- General queries → Direct response

---
### 🛠️ What You Need to Implement
- Agent logic
- Conditional routing
- Tool integration
- Basic error handling

### 🚀 Bonus
- Improve routing
- Add logging
- Add more tools


````python
import logging

logging.basicConfig(
    level=logging.INFO,
    format="%(asctime)s - %(message)s"
)
````

````python
# 🛠️ TOOL 1: Calculator

def calculator(expression: str) -> str:
    """Evaluate a mathematical expression."""
    try:
        return str(eval(expression))
    except Exception:
        return "Error in calculation"
````

````python
# 🛠️ TOOL 2: Keyword Extractor

def extract_keywords(text: str) -> list:
    """Extract keywords from text."""
    try:
        words = text.split()
        keywords = list(set([w.lower() for w in words if len(w) > 4]))
        return keywords[:5]
    except Exception:
        return []
````

````python
#  🛠️ TOOL 3: Word Counter
def word_counter(text):
    return len(text.split())
````

````python
# 🛠️ TOOL 4: Text Reverser
def reverse_text(text):
    return text[::-1]
````

## 🤖 Implement Agent Logic Below

👉 Use conditional routing:
- If query contains "calculate" → use calculator
- If query contains "keywords" → use keyword extractor
- Else → general response

````python
# 🤖 AGENT FUNCTION (TO IMPLEMENT)

def agent(query: str):
    query_lower = query.lower()
    logging.info(f"Received Query: {query}")
    try:

        # For Calculator routing

        math_words = [
          "calculate",
          "+",
          "-",
          "*",
          "/"
      ]

     # for keyboard routing
        keyword_words = [
        "keyword",
        "keywords",
        "extract"
     ]

        if any(word in query_lower for word in math_words):
            logging.info("Routing to Calculator Tool")
            expression = query_lower.replace("calculate", "").strip()

            answer = calculator(expression)

            return {
                "type": "calculation",
                "result": answer
            }

        # Keyword routing
        elif any(word in query_lower for word in keyword_words):
            logging.info("Routing to Keyword Tool")
            text = query.replace("Extract keywords from", "")
            text = text.replace("extract keywords from", "").strip()

            keywords = extract_keywords(text)

            return {
                "type": "keywords",
                "result": keywords
            }


        elif "count words" in query_lower:

          text = query_lower.replace("count words", "").strip()

          return {
              "type": "word_count",
              "result": word_counter(text)
           }

        elif "reverse" in query_lower:

          text = query.replace("reverse","").strip()

          return {
              "type":"reverse",
              "result": reverse_text(text)
           }

        # General response
        else:
            logging.info("General Query")
            return {
                "type": "general",
                "result": "This is a general query. No external tool required."
            }

    except Exception as e:
        logging.exception("Agent Error")
        return {
            "type": "error",
            "result": str(e)
        }

    return {
        "type": "unknown",
        "result": "Not implemented"
    }
````

## 📦 Expected Output Format

```
{
  "type": "calculation / keywords /count words / reverse/ general / error",
  "result": ...
}
```

````python
# 🧪 Test Cases

queries = [
    "Calculate 20 + 5",
    "Extract keywords from Artificial Intelligence is transforming industries",
    "What is machine learning?",
    "Count words Artificial Intelligence is amazing",
    "reverse Hello World"
]

for q in queries:
    print("Query:", q)
    print("Response:", agent(q))
    print("-" * 50)
````

````python
# 🎯 Interactive Mode

while True:
    user_input = input("Enter query (type 'exit' to stop): ")
    if user_input.lower() == "exit":
        break
    print("Response:", agent(user_input))
````
