### 1.2 Data Structures

Given the list `words = ['apple', 'banana', 'cherry', 'apple', 'date', 'banana', 'apple']`:

1. Build a **dict** `word_count` mapping each word to its frequency using only a loop (no `Counter`).
2. Using a **set**, find all unique words and store in `unique_words`.
3. Using a **list comprehension**, build `long_words` containing words with more than 5 characters.

```python
words = ['apple', 'banana', 'cherry', 'apple', 'date', 'banana', 'apple']

# 1. word_count dict — loop through and count each occurrence
word_count = {}
for w in words:
    word_count[w] = word_count.get(w, 0) + 1

# 2. unique_words — a set automatically removes duplicates
unique_words = set(words)

# 3. long_words — list comprehension filtering by length
long_words = [w for w in words if len(w) > 5]

# Tests
assert word_count == {'apple': 3, 'banana': 2, 'cherry': 1, 'date': 1}
assert unique_words == {'apple', 'banana', 'cherry', 'date'}
assert set(long_words) == {'banana', 'cherry'}
print("1.2 passed ✓")
```

### 1.3 Exceptions

Write `safe_divide(a, b)` that:
- Returns `a / b` normally
- Returns `None` if `b == 0` (catch `ZeroDivisionError`)
- Raises `TypeError` with message `"Inputs must be numeric"` if either input is not a number

```python
def safe_divide(a, b):
    # check types first before attempting division
    if not isinstance(a, (int, float)) or not isinstance(b, (int, float)):
        raise TypeError("Inputs must be numeric")
    try:
        return a / b
    except ZeroDivisionError:
        return None

# Tests
assert safe_divide(10, 2) == 5.0
assert safe_divide(5, 0) is None
try:
    safe_divide('x', 2)
    assert False, "Should have raised TypeError"
except TypeError as e:
    assert str(e) == "Inputs must be numeric"
print("1.3 passed ✓")
```

### 1.4 Functions & Lambdas

2. Using a **lambda**, create `triple` that multiplies its input by 3.
3. Demonstrate: `apply_twice(triple, 4)` should return `36`.

```python
def apply_twice(f, x):
    # apply f to x, then apply f to that result
    return f(f(x))

triple = lambda x: x * 3  # anonymous function: multiply by 3

result = apply_twice(triple, 4)  # triple(4)=12, triple(12)=36
assert result == 36, f"Expected 36, got {result}"
print("1.4 passed ✓")
```

### 4.2 Matrix Operations in Practice

```python
P = np.array([[2, 1], [0, 3]])
Q = np.array([[1, 4], [2, 0]])

PplusQ   = P + Q      # element-wise addition
scalar3P = 3 * P      # multiply every element by 3
PQ       = P @ Q      # matrix multiplication: (P rows) dot (Q cols)
QP       = Q @ P      # reversed order — generally different result

print(f"P + Q =\n{PplusQ}")
print(f"3*P =\n{scalar3P}")
print(f"P @ Q =\n{PQ}")
print(f"Q @ P =\n{QP}")
print(f"PQ == QP? {np.array_equal(PQ, QP)}")

assert not np.array_equal(PQ, QP), "They should differ!"
print("4.2 passed ✓")
```

### 4.3 Eigenvalues & Eigenvectors

```python
A = np.array([[4, 1], [2, 3]], dtype=float)

# eig returns eigenvalues (λ) and a matrix whose columns are eigenvectors
eigenvalues, eigenvectors = np.linalg.eig(A)

print(f"Eigenvalues: {eigenvalues}")
print(f"Eigenvectors (columns):\n{eigenvectors}")

# Verify the eigenvector equation: A @ v == λ * v for each pair
for i in range(len(eigenvalues)):
    lam = eigenvalues[i]
    vec = eigenvectors[:, i]
    lhs = A @ vec
    rhs = lam * vec
    print(f"  λ={lam:.1f}: A@v={lhs}, λ*v={rhs}, match={np.allclose(lhs, rhs)}")

# Plot eigenvectors as scaled arrows
fig, ax = plt.subplots(figsize=(5, 5))
colors = ['steelblue', 'tomato']
for i in range(2):
    vec = eigenvectors[:, i] * eigenvalues[i]
    ax.quiver(0, 0, vec[0], vec[1], angles='xy', scale_units='xy', scale=1,
              color=colors[i], width=0.04, label=f'λ={eigenvalues[i]:.0f}')
ax.set_xlim(-4, 4)
ax.set_ylim(-4, 4)
ax.axhline(0, color='k', lw=0.5)
ax.axvline(0, color='k', lw=0.5)
ax.set_title('Eigenvectors (scaled by eigenvalue)')
ax.legend()
plt.tight_layout()
plt.show()

print("4.3 passed ✓")
```

**Your Geometric Explanation (4.3):**

> An eigenvector of a matrix is a special direction that doesn't get rotated by the transformation — it only gets stretched or squished. The eigenvalue tells you by how much: if λ > 1 the vector is stretched, if 0 < λ < 1 it is compressed, and if λ < 0 it gets flipped. Most input vectors get both rotated and scaled when multiplied by a matrix, but eigenvectors are the rare directions that stay perfectly aligned — they just change in length.

### 4.4 SVD & Dimensionality Reduction

```python
np.random.seed(42)
X = np.random.randn(4, 3)

# SVD decomposes X into three matrices: X = U @ diag(S) @ Vt
U, S, Vt = np.linalg.svd(X, full_matrices=False)

# Reconstruct by reversing the decomposition
X_reconstructed = U @ np.diag(S) @ Vt

# Rank-1 approximation: keep only the largest singular value and its vectors
X_approx = S[0] * np.outer(U[:, 0], Vt[0, :])

print(f"U shape: {U.shape}, S shape: {S.shape}, Vt shape: {Vt.shape}")
print(f"Singular values: {S}")
print(f"Reconstruction error: {np.linalg.norm(X - X_reconstructed):.2e}")
print(f"Rank-1 approximation:\n{X_approx}")

assert np.allclose(X, X_reconstructed, atol=1e-10), "Reconstruction failed!"
print("4.4 passed ✓")
```

**SVD → PCA Connection (4.4):**

> In the SVD decomposition X = U S Vᵀ, the columns of **Vᵀ** (rows of Vt) are the principal directions — the directions of maximum variance in the data. The singular values in **S** encode how much variance each direction captures (they are proportional to the square root of PCA eigenvalues). The columns of **U** are the coordinates of each data point projected onto those directions. PCA is essentially SVD applied to the mean-centred data matrix: the right singular vectors (Vt rows) are exactly the principal components.

### 5.3 Error Metrics

```python
y_true = np.array([3.0, 5.0, 2.5, 7.0, 4.5, 6.0, 1.5, 8.0])
y_pred = np.array([2.8, 5.2, 2.1, 7.5, 4.0, 6.3, 2.0, 7.8])
n, p = len(y_true), 2

mae    = np.mean(np.abs(y_true - y_pred))           # average absolute error
mse    = np.mean((y_true - y_pred) ** 2)            # penalises large errors more
rmse   = np.sqrt(mse)                               # same units as target

# R² = 1 - (residual sum of squares / total sum of squares)
ss_res = np.sum((y_true - y_pred) ** 2)
ss_tot = np.sum((y_true - y_true.mean()) ** 2)
r2     = 1 - ss_res / ss_tot

# Adjusted R² penalises adding more predictors that don't help
adj_r2 = 1 - (1 - r2) * (n - 1) / (n - p - 1)

print(f"MAE:        {mae:.4f}")
print(f"MSE:        {mse:.4f}")
print(f"RMSE:       {rmse:.4f}")
print(f"R²:         {r2:.4f}")
print(f"Adj. R²:    {adj_r2:.4f}")
```

### 5.4 Distribution Testing & Stationarity

```python
from statsmodels.tsa.stattools import adfuller

np.random.seed(0)
s1 = np.random.normal(0, 1, 200)
s2 = np.random.exponential(1, 200)

# KS test — compare each sample against a standard normal CDF
ks_s1 = stats.kstest(s1, 'norm')
ks_s2 = stats.kstest(s2, 'norm')
print(f"KS test s1 (normal):      stat={ks_s1.statistic:.4f}, p={ks_s1.pvalue:.4f}")
print(f"KS test s2 (exponential): stat={ks_s2.statistic:.4f}, p={ks_s2.pvalue:.4f}")

# Create a non-stationary time series (linear trend + noise)
t  = np.arange(200)
ts = 0.05 * t + np.random.normal(0, 1, 200)

adf_result = adfuller(ts)
print(f"\nADF statistic: {adf_result[0]:.4f}")
print(f"ADF p-value:   {adf_result[1]:.4f}")
print(f"Stationary?    {adf_result[1] < 0.05}")

# First-order differencing removes the linear trend
ts_diff  = np.diff(ts)
adf_diff = adfuller(ts_diff)
print(f"\nAfter differencing — p-value: {adf_diff[1]:.4f}, Stationary? {adf_diff[1] < 0.05}")
```

### 5.5 Model Monitoring Concepts

```python
def compute_psi(expected, actual, bins=10):
    """
    PSI = sum((actual_pct - expected_pct) * ln(actual_pct / expected_pct))
    Add epsilon=1e-10 to avoid log(0).
    """
    # build bin edges from percentiles of the expected distribution
    breaks = np.percentile(expected, np.linspace(0, 100, bins + 1))
    breaks[0]  = -np.inf   # make sure every value fits in the leftmost bin
    breaks[-1] =  np.inf   # and every value fits in the rightmost bin

    eps = 1e-10  # small constant to prevent log(0)

    exp_counts = np.histogram(expected, bins=breaks)[0]
    act_counts = np.histogram(actual,   bins=breaks)[0]

    exp_pct = exp_counts / len(expected) + eps
    act_pct = act_counts / len(actual)   + eps

    psi = np.sum((act_pct - exp_pct) * np.log(act_pct / exp_pct))
    return psi


np.random.seed(1)
train_dist = np.random.normal(50, 10, 1000)   # training distribution
drift_dist = np.random.normal(65, 12, 1000)   # shifted production distribution

psi_value = compute_psi(train_dist, drift_dist)
print(f"PSI: {psi_value:.4f}")
print(f"Shift severity: {'Major' if psi_value > 0.2 else 'Minor' if psi_value > 0.1 else 'Stable'}")

# Plot both distributions to visualise the shift
fig, ax = plt.subplots()
ax.hist(train_dist, bins=40, density=True, alpha=0.5,
        color='steelblue', edgecolor='white', label='Training (expected)')
ax.hist(drift_dist, bins=40, density=True, alpha=0.5,
        color='tomato', edgecolor='white', label='Production (actual)')
ax.legend()
ax.set_xlabel('Feature value')
ax.set_title('Distribution Shift Visualization')
plt.tight_layout()
plt.show()
```

**Concept Drift vs Covariate Drift & PSI Thresholds (5.5):**

- **Concept drift:** The statistical relationship between the input features and the target variable changes over time — the model's learned mapping becomes incorrect even if the input distribution stays the same.
- **Covariate drift:** The distribution of the input features themselves shifts, while the underlying input-to-output relationship remains the same — the model may still be conceptually correct but receives out-of-distribution inputs.
- **PSI < 0.1:** No significant shift; the production distribution closely matches training and no action is needed.
- **PSI 0.1–0.2:** Minor shift detected; monitor closely and investigate which features are drifting.
- **PSI > 0.2:** Major shift; the model is likely degraded and retraining or recalibration should be triggered.
- **Retraining trigger example:** Schedule automatic retraining when PSI exceeds 0.2 on any of the top-5 most important features for two consecutive monitoring windows.

## Part 6 — Probability Theory

### 6.1 Core Concepts

```python
total = 10
red, blue, green = 4, 3, 3

p_red   = red   / total   # 4/10
p_blue  = blue  / total   # 3/10
p_green = green / total   # 3/10

# Joint: P(red first AND blue second) — without replacement
# = P(red first) * P(blue second | red first)
p_red_then_blue  = (red / total) * (blue / (total - 1))

# Conditional: given red was drawn, 9 marbles remain, 3 are blue
p_blue_given_red = blue / (total - 1)

# Independence check: draws are independent iff P(B|A) == P(B)
# Here P(blue|red) = 3/9 ≈ 0.333 ≠ P(blue) = 3/10 = 0.3, so NOT independent
independent = (p_blue_given_red == p_blue)

print(f"P(red)={p_red:.3f}, P(blue)={p_blue:.3f}, P(green)={p_green:.3f}")
print(f"P(red,blue): {p_red_then_blue:.4f}")
print(f"P(blue|red): {p_blue_given_red:.4f}")
print(f"P(blue):     {p_blue:.4f}")
print(f"Independent? {independent}")
```

### 6.3 Bayes' Theorem

```python
p_spam            = 0.30
p_free_given_spam = 0.80
p_free_given_ham  = 0.05
p_ham             = 1 - p_spam

# P(free) via the law of total probability
p_free = p_free_given_spam * p_spam + p_free_given_ham * p_ham

# Bayes: P(Spam | free) = P(free | Spam) * P(Spam) / P(free)
p_spam_given_free = (p_free_given_spam * p_spam) / p_free

print(f"P('free'): {p_free:.4f}")
print(f"P(Spam | 'free'): {p_spam_given_free:.4f}")


def naive_bayes_predict(prior_spam, p_word_given_spam, p_word_given_ham):
    """Return posterior P(Spam | word) using Bayes' theorem."""
    prior_ham = 1 - prior_spam
    # evidence: total probability of seeing this word
    p_word = p_word_given_spam * prior_spam + p_word_given_ham * prior_ham
    return (p_word_given_spam * prior_spam) / p_word


pred = naive_bayes_predict(0.30, 0.80, 0.05)
assert abs(pred - p_spam_given_free) < 1e-9
print(f"naive_bayes_predict: {pred:.4f}")
print("6.3 passed ✓")
```

**Bayes Term Mapping (6.3):**

| Term | Formula | In this example |
|------|---------|-----------------|
| Prior | P(Spam) | Our initial belief that any email is spam: 0.30 |
| Likelihood | P(word \| Spam) | Probability of seeing the word "free" given the email is spam: 0.80 |
| Evidence | P(word) | Overall probability of any email containing "free": 0.275 |
| Posterior | P(Spam \| word) | Updated belief that the email is spam after observing "free": ≈ 0.873 |

### 6.4 Central Limit Theorem

```python
np.random.seed(7)
population = np.random.exponential(scale=1.0, size=100_000)

n_samples   = 5000
sample_size = 30

# draw 5000 random samples and record each sample's mean
sample_means = np.array([
    np.random.choice(population, size=sample_size, replace=False).mean()
    for _ in range(n_samples)
])

pop_mean = population.mean()
pop_std  = population.std()
clt_std  = pop_std / np.sqrt(sample_size)   # standard error of the mean

print(f"Population mean: {pop_mean:.4f}, Population std: {pop_std:.4f}")
print(f"Sample means mean: {sample_means.mean():.4f}, std: {sample_means.std():.4f}")
print(f"CLT predicted std: {clt_std:.4f}")

# Histogram of sample means + theoretical normal overlay
fig, ax = plt.subplots()
ax.hist(sample_means, bins=50, density=True, alpha=0.6,
        color='steelblue', edgecolor='white', label='Sample means')
x_range = np.linspace(sample_means.min(), sample_means.max(), 200)
ax.plot(x_range, stats.norm.pdf(x_range, pop_mean, clt_std),
        color='tomato', lw=2, label='CLT normal')
ax.set_xlabel('Sample mean')
ax.set_title('CLT: Distribution of Sample Means')
ax.legend()
plt.tight_layout()
plt.show()

# KS test to confirm approximate normality
ks_result = stats.kstest(sample_means, 'norm', args=(pop_mean, clt_std))
print(f"KS test p-value: {ks_result.pvalue:.4f} → Approximately normal? {ks_result.pvalue > 0.05}")
```

**CLT Reflection (6.4):**

> The CLT tells us that regardless of how non-normal the underlying population is (here it is exponential), the distribution of sample means will be approximately normal once the sample size is large enough. This is foundational for statistics and ML because it justifies using tools like t-tests, confidence intervals, and z-scores even when we cannot assume the raw data is normally distributed. Many classical tests implicitly rely on the CLT to guarantee that test statistics converge to known distributions as sample sizes grow.
